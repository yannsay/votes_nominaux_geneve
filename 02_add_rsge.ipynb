```python
import pandas as pd
```

```python
votings = pd.read_csv("inputs/voting_body.csv")
```

```python
column_missing :str = []
for column in votings.columns:
    if all(pd.isna(votings[column])):
        column_missing.append(column)

column_missing
voting_clean = votings.drop(columns = column_missing)
voting_clean
    
```

```python
clean_rsge = pd.read_csv("outputs/reformatted_rsge.csv")
clean_rsge
```

```python
voting_clean["voting_affair_title_fr"]
```

```python
voting_clean["voting_affair_title_fr"][0]
```

```python
pattern = r'([A-Z] [0-9] [0-9]{2}|[A-Z]\xa0[0-9]\xa0[0-9]{2})'
voting_clean['Référence'] = voting_clean['voting_affair_title_fr'].str.extract(pattern)
```

```python
sum(pd.isna(voting_clean['Référence']))/(voting_clean.shape[0])
```

```python
voting_clean.loc[~pd.isna(voting_clean['Référence']),]
```

```python
pattern = r'([A-Z]{1,2})'
voting_clean['type_vote'] = voting_clean['voting_affair_number'].str.extract(pattern)
```

```python
voting_clean['type_vote']
```

```python
voting_clean.loc[~pd.isna(voting_clean['Référence']),"type_vote"].value_counts()
```

```python
voting_clean.loc[pd.isna(voting_clean['Référence']) & (voting_clean["type_vote"] != "PL"),]
```

```python
voting_clean.columns
```

```python
pl_voting = voting_clean.loc[~pd.isna(voting_clean['Référence']),].reset_index()
pl_voting = pl_voting.drop(columns = ["index", "voting_body_key", "voting_updated_local", "voting_created_local"])
pl_voting
```

```python
pl_voting['Référence'] = pl_voting['Référence'].str.replace('\xa0', ' ', regex=False)
```

```python
pl_voting= pl_voting.merge(clean_rsge, on = "Référence", how = "left")
```

```python
pl_voting.to_csv("outputs/pl_voting_clean.csv", index=False)
```
